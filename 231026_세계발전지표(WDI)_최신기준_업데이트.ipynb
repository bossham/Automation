```python
year_columns = filtered_df.columns[2:]

# 각 국가의 최신 연도와 값을 저장할 리스트
latest_data = []

for index, row in filtered_df.iterrows():
    latest_year = row[year_columns].last_valid_index()  # 가장 최근 연도
    latest_value = row[latest_year] if latest_year else None  # 해당 연도의 값 (또는 None)
    latest_data.append([row['국가코드3자리'], row['국가한글명'], latest_year, latest_value])

latest_data_df = pd.DataFrame(latest_data, columns=['국가코드3자리', '국가한글명', '가장 최근 연도', '해당 연도의 값'])

## 수정
# # NaN 값을 포함하는 행 제외
# latest_data_df = latest_data_df.dropna()

latest_data_df = latest_data_df.sort_values(by='국가한글명')

print(latest_data_df.shape)
latest_data_df
```
